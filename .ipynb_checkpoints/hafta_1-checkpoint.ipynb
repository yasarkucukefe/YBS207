# Big O Notasyonu

```python
# Bir algoritmanın komplekslik derecesini gösteren bir ölçüttür.
# O(n)
# O(n^2)
# O(1)
# O(log n)
# O(2^n)
```

```python
# O(n) 
liste = [1,2,3,4,5,6,7]
for rakam in liste: # Algoritma
    print(rakam)
```

```python
# O(n^2)
liste1 = [1,2,3,4]
liste2 = [4,5,6,7]
for rakam1 in liste1:
    for rakam2 in liste2:
        print(rakam1,rakam2)
```

```python
# O(1)
liste = [1,2,3,4]
liste.append(5)
```

```python
# O(log n)
# İlk örnekteki sayı tahmin algoritması O(log n)
```

```python
# O(2^n)
# Fibonacci serisi 0,1,1,2,3,5
def fibonacci(number):
    if number <= 1:
        return number
    return fibonacci(number - 2) + fibonacci(number - 1)
```

```python
fibonacci(10) # >100 değerleri denemeyin !
```

### Bir listedeki en küçük elemanın bulunması

```python
# O(n)
liste = [3,9,8,11,7]
en_kucuk = 1000
for rakam in liste:
    if rakam < en_kucuk:
        en_kucuk = rakam
print(en_kucuk)
```

```python
class Kitap:
    yazar_isim = ""
    sayfa_say = 0
    def __init__(self,yazar,sayfa):
        self.yazar_isim = yazar
        self.sayfa_say = sayfa
        
    def __str__(self):
        return f"Yazar adı: {self.yazar_isim}, sayfa sayısı: {self.sayfa_say}"
```

```python
kitap3 = Kitap("Garcia Marquez",300)
```

```python
print(kitap3)
```
